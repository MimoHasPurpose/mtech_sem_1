```python
from numpy.random import rand
from numpy import hstack
from matplotlib import pyplot
```

```python
def generate_samples(n=100):
    X1=rand(n)-0.5
    X2=X1*X1
    X1=X1.reshape(n,1)
    X2=X2.reshape(n,1)
    return hstack((X1, X2))
```

```python
data=generate_samples()
```

```python
data.shape
```

```python
pyplot.scatter(data[:,0],data[:,1])
pyplot.show()
```
